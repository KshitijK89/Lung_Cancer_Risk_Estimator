```python
# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 2. Load Dataset
data_path = os.path.join("..", "data", "survey lung cancer.csv")
df = pd.read_csv(data_path)

# 3. Basic Info
print("âœ… Shape:", df.shape)
print("âœ… Columns:", df.columns.tolist())
print("\nðŸ§¾ Data Info:")
print(df.info())

# 4. Clean Column Names
df.columns = df.columns.str.strip().str.replace(" ", "_")

# 5. View First 5 Rows
print("\nðŸ‘€ Head:")
print(df.head())

# 6. Check Missing Values
print("\nðŸ©º Missing Values:")
print(df.isnull().sum())

# 7. Target Class Distribution
print("\nðŸŽ¯ Class Balance:")
print(df['LUNG_CANCER'].value_counts())

# 8. Visualize Class Balance
sns.countplot(x='LUNG_CANCER', data=df)
plt.title("Lung Cancer Class Distribution")
plt.savefig('../outputs/charts/class_distribution.png')
plt.show()

# 9. Correlation Heatmap (for numerical features)
numerical = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10,6))
sns.heatmap(numerical.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.savefig('../outputs/charts/correlation_heatmap.png')
plt.show()

# 10. Gender vs Cancer Risk
sns.countplot(x='GENDER', hue='LUNG_CANCER', data=df)
plt.title("Lung Cancer by Gender")
plt.savefig('../outputs/charts/gender_vs_cancer.png')
plt.show()
```
